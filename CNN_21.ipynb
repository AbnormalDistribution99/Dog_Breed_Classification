```python
import numpy as np 
import tensorflow as tf
from tensorflow.keras.layers import Conv2D, MaxPool2D, Dense, Flatten
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Sequential
from PIL import Image
import os

import warnings
warnings.filterwarnings("ignore")
```

```python
train_dir = "/Users/abhishekregmi/Desktop/Sandeep HW/CNN_dataset/Training"
test_dir = "/Users/abhishekregmi/Desktop/Sandeep HW/CNN_dataset/Test"
```

```python
number_of_classes = 20
classifier = Sequential()
classifier.add(Conv2D(20, (3,3), input_shape = (64,64,3), activation = 'relu')) # 20 * 3 * 3 ==>  
classifier.add(Conv2D(20, (3,3), padding = "same", activation = 'relu'))
classifier.add(Conv2D(20, (3,3), padding = "same", activation = 'relu'))
classifier.add(MaxPool2D(pool_size = (2,2), padding = "same"))
classifier.add(Conv2D(40, (3,3), activation = 'relu'))
classifier.add(Conv2D(40, (3,3), activation = 'relu'))
classifier.add(Conv2D(40, (3,3), activation = 'relu'))
classifier.add(Flatten())
classifier.add(Dense(40, activation = 'relu'))
classifier.add(Dense(20, activation = 'softmax'))
```

```python
OPTIMIZER = tf.optimizers.SGD(learning_rate=0.01)
LOSS = tf.losses.categorical_crossentropy
METRICS = ['accuracy']
```

```python
classifier.compile(
    optimizer = OPTIMIZER,
    loss = LOSS,
    metrics= METRICS
)
```

```python
train_datagen = ImageDataGenerator(
    rescale = 1./255,
    rotation_range=40,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest',
    validation_split= 0.2,
)
```

```python
test_datagen = ImageDataGenerator(
    rescale = 1./255)
```

```python
os.getcwd()
```

```python
training_set = train_datagen.flow_from_directory(
    train_dir,
    target_size = (64,64),
    batch_size = 16
)
```

```python
test_set = test_datagen.flow_from_directory(
    test_dir,
    target_size = (64, 64),
    batch_size = 16
)
```
